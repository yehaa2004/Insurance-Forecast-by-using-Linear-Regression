```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

# 1. Load dataset (download from Kaggle: insurance.csv)
df = pd.read_csv("/content/insurance.csv")

# 2. Features and target
X = df.drop("charges", axis=1)
y = df["charges"]

# 3. Separate numeric and categorical features
numeric_features = ["age", "bmi", "children"]
categorical_features = ["sex", "smoker", "region"]

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])

# 4. Define models
models = {
    "Linear Regression": Pipeline([
        ("pre", preprocessor),
        ("model", LinearRegression())
    ]),
    "Ridge Regression": Pipeline([
        ("pre", preprocessor),
        ("model", Ridge(alpha=10))
    ]),
    "Lasso Regression": Pipeline([
        ("pre", preprocessor),
        ("model", Lasso(alpha=0.1))
    ])
}

# 5. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Train and evaluate
print("\nModel Performance:\n")
for name, pipeline in models.items():
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name}: MSE = {mse:.2f}, R² = {r2:.3f}")

# 7. Learning curve (for Linear Regression as an example)
train_sizes, train_scores, test_scores = learning_curve(
    models["Linear Regression"], X_train, y_train, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 5), scoring='neg_mean_squared_error'
)

train_mse = -np.mean(train_scores, axis=1)
test_mse = -np.mean(test_scores, axis=1)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_mse, 'o-', label='Training MSE')
plt.plot(train_sizes, test_mse, 's-', label='Validation MSE')
plt.xlabel('Training set size')
plt.ylabel('MSE')
plt.title('Learning Curve: Linear Regression (Insurance Dataset)')
plt.legend()
plt.grid(True)
plt.show()
```
